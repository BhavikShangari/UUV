```python
enhanced = pd.read_csv('enhanced_results.csv')
unenhanced = pd.read_csv('unenhanced_results.csv')
```

```python
enhanced.columns
```

# Train/Box loss

```python
plt.plot(enhanced['                  epoch'], enhanced['         train/box_loss'])
plt.plot(unenhanced['                  epoch'], unenhanced['         train/box_loss'])
plt.legend(['enhanced', 'unenhanced'])
plt.ylabel('train/box_loss')
```

# Train/cls loss

```python
plt.plot(enhanced['                  epoch'], enhanced['         train/cls_loss'])
plt.plot(unenhanced['                  epoch'], unenhanced['         train/cls_loss'])
plt.legend(['enhanced', 'unenhanced'])
plt.ylabel('train/cls_loss')
```
