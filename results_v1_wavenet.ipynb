```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
unenhanced = pd.read_csv('/home/bhavik/UUV/runs/detect/nano_unenhanced_v1/results.csv')
enhanced = pd.read_csv('/home/bhavik/UUV/runs/detect/nano_enhanced_v1/results.csv')
```

```python
unenhanced.columns
```
